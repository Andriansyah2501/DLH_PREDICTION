```python
import pandas as pd
import numpy as np

file_name = 'Test.xls'

# ==========================================
# 1. MEMBACA & MENYIAPKAN REFERENSI ARMADA
# ==========================================
print(" Membaca referensi List Armada...")
df_ref = pd.read_excel(file_name, sheet_name='List Armada', header=1)

# Bersihkan spasi dan standarkan format teks untuk pencocokan data
df_ref['NOPIN'] = df_ref['NOPIN'].astype(str).str.strip().str.upper()
df_ref['NO.PLAT'] = df_ref['NO.PLAT'].astype(str).str.strip().str.upper()

# Konversi referensi menjadi dictionary agar proses pencarian cepat
ref_dict = df_ref.set_index('NOPIN')[['NO.PLAT', 'Lokasi KECAMATAN', 'MERK', 'TYPE']].to_dict('index')
print("✔ Referensi 'List Armada' berhasil dimuat.\n")

# ==========================================
# 2. PROSES MEMBACA SELURUH SHEET (1-30) DIALAKUKAN SECARA DINAMIS
# ==========================================
cleaned_sheets = {}
total_baris_bersih = 0

# Mengambil semua nama sheet yang berupa angka (1 sampai 30)
xl = pd.ExcelFile(file_name)
hari_sheets = [sheet for sheet in xl.sheet_names if sheet.isdigit()]

print(f" Memulai pemindaian otomatis untuk {len(hari_sheets)} sheet harian...")

for hari in hari_sheets:
    # Langkah A: Baca secara mentah (tanpa berasumsi baris header ada di mana)
    df_raw = pd.read_excel(file_name, sheet_name=hari, header=None)

    # Langkah B: Cari baris mana yang merupakan baris nama kolom (header)
    header_idx = None
    for idx, row in df_raw.iterrows():
        # Gabungkan teks dalam satu baris untuk mendeteksi kata kunci
        row_str = " ".join(row.astype(str).dropna().str.upper().values)
        if 'PINTU' in row_str or 'PLAT MOBIL' in row_str:
            header_idx = idx
            break

    if header_idx is None:
        print(f"⚠ Sheet {hari}: Lewat (Tidak ditemukan baris header nama kolom).")
        continue

    # Langkah C: Baca ulang sheet dengan baris header yang tepat yang sudah ditemukan
    df_hari = pd.read_excel(file_name, sheet_name=hari, header=header_idx)

    # Langkah D: Bersihkan dan standarkan nama kolom agar seragam (Upper & No Space)
    df_hari.columns = [str(col).strip().upper() for col in df_hari.columns]

    # Langkah E: Temukan kolom No. Pintu dan No. Plat secara fleksibel (karena penulisan sering berbeda)
    col_nopin = [c for c in df_hari.columns if 'PINTU' in c or 'NOPIN' in c]
    col_plat = [c for c in df_hari.columns if 'PLAT' in c]

    if not col_nopin or not col_plat:
        print(f"⚠ Sheet {hari}: Lewat (Kolom identitas armada tidak lengkap).")
        continue

    # Ganti nama kolom ke format standar internal kita
    df_hari.rename(columns={col_nopin[0]: 'NOPIN', col_plat[0]: 'NO_PLAT'}, inplace=True)

    # ==========================================
    # 3. FILTERING & PEMBERSIHAN BARIS DATA
    # ==========================================
    # 1. Hapus jika kolom NOPIN kosong
    df_hari = df_hari.dropna(subset=['NOPIN'])
    df_hari['NOPIN'] = df_hari['NOPIN'].astype(str).str.strip().str.upper()

    # 2. Singkirkan baris summary bawah seperti Total, Goro, Jumlah, dll.
    df_hari = df_hari[~df_hari['NOPIN'].str.contains('TOTAL|GORO|JUMLAH|KETERANGAN|NAN|COLUMN', na=False)]
    df_hari = df_hari[df_hari['NOPIN'] != '']

    # 3. Standardisasi float tidak sengaja dari excel (misal '71.0' menjadi '71')
    df_hari['NOPIN'] = df_hari['NOPIN'].apply(lambda x: x[:-2] if x.endswith('.0') else x)

    # ==========================================
    # 4. EDIT & SINKRONISASI DATA BERDASARKAN MASTER
    # ==========================================
    def sinkronisasi_master(row):
        nopin = row['NOPIN']
        if nopin in ref_dict:
            # Jika No Pintu terdaftar di List Armada, paksa ikuti data List Armada
            row['NO_PLAT'] = ref_dict[nopin]['NO.PLAT']
            row['Kecamatan'] = ref_dict[nopin]['Lokasi KECAMATAN']
            row['MERK'] = ref_dict[nopin]['MERK']
            row['TYPE'] = ref_dict[nopin]['TYPE']
        return row

    # Jalankan sinkronisasi data
    df_hari = df_hari.apply(sinkronisasi_master, axis=1)

    # Tambahkan kolom penanda Tanggal (asumsi Juni 2026 berdasarkan data rekap Anda)
    df_hari['TANGGAL'] = f"2026-06-{int(hari):02d}"

    # Simpan hasil sementara
    cleaned_sheets[hari] = df_hari
    total_baris_bersih += len(df_hari)

print(f"\n✔ Berhasil memproses total {len(cleaned_sheets)} lembar kerja harian!")
print(f"✔ Total data valid terkumpul: {total_baris_bersih} baris.")

# ==========================================
# 5. BERGABUNG MENJADI SATU MASTER DATA UTUH
# ==========================================
if cleaned_sheets:
    df_master_total = pd.concat(cleaned_sheets.values(), ignore_index=True)
    print("✔ Seluruh data berhasil digabungkan menjadi Master Data tunggal!")

    # Tampilkan preview sampel data gabungan
    print("\n=== PREVIEW 5 DATA PERTAMA GABUNGAN ===")
    kolom_preview = ['TANGGAL', 'NOPIN', 'NO_PLAT', 'Kecamatan', 'MERK', 'TYPE']
    display(df_master_total[[c for c in kolom_preview if c in df_master_total.columns]].head())
```

```python
# =====================================================================
# JAWABAN POIN 2.0: KATEGORISASI DATA BERDASARKAN WILAYAH KECAMATAN
# =====================================================================

print("=== PROSES PENYUSUNAN JAWABAN POIN 2.0 ===\n")

# 1. Pastikan kolom berat bersih (NETTO) bertipe numerik agar bisa dijumlahkan
# Jika di sheet Anda namanya 'NETTO', 'TOTAL NETTO', atau 'BERAT', kita sesuaikan.
kolom_netto = [c for c in df_master_total.columns if 'NETTO' in c or 'TOTAL' in c and 'TRIP' not in c]
if kolom_netto:
    target_netto = kolom_netto[0]
    df_master_total[target_netto] = pd.to_numeric(df_master_total[target_netto], errors='coerce').fillna(0)
else:
    # Jika tidak ada kolom Netto spesifik, kita buat dummy/asumsi kolom ke-7 untuk demo kalkulasi
    target_netto = 'NETTO_ESTIMASI'
    df_master_total[target_netto] = 0

# 2. AGREGASI 1: Distribusi Log Armada & Estimasi Tonase per Kecamatan
df_kecamatan = df_master_total.groupby('Kecamatan').agg(
    Jumlah_Armada_Aktif=('NOPIN', 'nunique'),
    Total_Ritase_Trip=('NOPIN', 'count'),
    Total_Tonase_Bersih=(target_netto, 'sum')
).reset_index()

# Urutkan berdasarkan wilayah dengan aktivitas tertinggi
df_kecamatan = df_kecamatan.sort_values(by='Total_Ritase_Trip', ascending=False).reset_index(drop=True)

print(">>> 1. Ringkasan Distribusi Kerja per Wilayah Kecamatan:")
display(df_kecamatan)

# 3. AGREGASI 2: Distribusi Berdasarkan Tipe/Merk Kendaraan
df_tipe = df_master_total.groupby(['TYPE', 'MERK']).agg(
    Total_Trip=('NOPIN', 'count'),
    Total_Tonase=(target_netto, 'sum')
).reset_index().sort_values(by='Total_Trip', ascending=False).reset_index(drop=True)

print("\n>>> 2. Ringkasan Distribusi Berdasarkan Tipe & Merk Armada:")
display(df_tipe)

# 4. EKSPOR HASIL NOMOR 2 KE SHEET BARU
with pd.ExcelWriter('Laporan_Analisis_Kecamatan_Poin2.xlsx') as writer:
    df_kecamatan.to_excel(writer, sheet_name='Per Kecamatan', index=False)
    df_tipe.to_excel(writer, sheet_name='Per Tipe Armada', index=False)

print("\n🎉 Berkas Jawaban Poin 2.0 Berhasil Disimpan: 'Laporan_Analisis_Kecamatan_Poin2.xlsx'")
```

3. Analisis Trip & Akumulasi Tonase

```python
# =====================================================================
# JAWABAN POIN 3.0: ANALISIS RITASE (TRIP) & AKUMULASI TONASE ARMADA
# =====================================================================

print("=== PROSES PENYUSUNAN JAWABAN POIN 3.0 ===\n")

# 1. Identifikasi nama kolom Tonase secara fleksibel (Gross, Tare, Netto)
col_gross = [c for c in df_master_total.columns if 'GROSS' in c or 'BRUTO' in c]
col_tare = [c for c in df_master_total.columns if 'TARE' in c or 'TARRA' in c]
col_netto = [c for c in df_master_total.columns if 'NETTO' in c]

# Pastikan nama kolom yang dipakai terstandar, jika tidak ketemu gunakan nama default
g_col = col_gross[0] if col_gross else 'GROSS'
t_col = col_tare[0] if col_tare else 'TARE'
n_col = col_netto[0] if col_netto else 'NETTO'

# 2. Konversi seluruh kolom tonase ke tipe data numerik (mengantisipasi string kosong / teks eror)
for col in [g_col, t_col, n_col]:
    if col in df_master_total.columns:
        df_master_total[col] = pd.to_numeric(df_master_total[col], errors='coerce').fillna(0)
    else:
        df_master_total[col] = 0

# 3. AGREGASI UTAMA: Hitung Total Trip dan Akumulasi Tonase per Armada
df_armada_performa = df_master_total.groupby(['NOPIN', 'NO_PLAT', 'Kecamatan', 'TYPE', 'MERK']).agg(
    Total_Trip=('NOPIN', 'count'),
    Total_Gross_Kg=(g_col, 'sum'),
    Total_Tare_Kg=(t_col, 'sum'),
    Total_Netto_Kg=(n_col, 'sum')
).reset_index()

# 4. Tambahkan kalkulasi konversi ke Ton untuk mempermudah pelaporan manajemen
df_armada_performa['Total_Netto_Ton'] = (df_armada_performa['Total_Netto_Kg'] / 1000).round(2)

# Urutkan dari armada yang memiliki performa tonase angkutan terbesar
df_armada_performa = df_armada_performa.sort_values(by='Total_Netto_Kg', ascending=False).reset_index(drop=True)

print(f"✔ Berhasil mengekstrak performa kuantitatif untuk {len(df_armada_performa)} unit armada aktif.")
print(">>> Ringkasan 10 Armada dengan Muatan Tonase Terbesar (Poin 3.0):")
display(df_armada_performa.head(10))

# 5. EKSPOR JAWABAN NOMOR 3 KE EXCEL
nama_file_3 = 'Laporan_Performa_Tonase_Armada_Poin3.xlsx'
df_armada_performa.to_excel(nama_file_3, index=False, sheet_name='Performa Tonase Juni 2026')

print(f"\n🎉 Berkas Jawaban Poin 3.0 Berhasil Diekspor: '{nama_file_3}'")
```

4. Tren Volume & Ritase Harian

```python
# =====================================================================
# JAWABAN POIN 4.0: ANALISIS TREN TONASE DAN RITASE HARIAN (JUNI 2026)
# =====================================================================

print("=== PROSES PENYUSUNAN JAWABAN POIN 4.0 ===\n")

# 1. Pastikan kolom NETTO standar yang kita gunakan sudah bersih
n_col = col_netto[0] if col_netto else 'NETTO'

# 2. AGREGASI HARIAN: Hitung Total Trip dan Total Tonase berdasarkan Tanggal
df_tren_harian = df_master_total.groupby('TANGGAL').agg(
    Total_Ritase_Trip=('NOPIN', 'count'),
    Total_Tonase_Kg=(n_col, 'sum'),
    Rata_Rata_Muatan_Per_Trip=(n_col, 'mean'),
    Jumlah_Armada_Operasional=('NOPIN', 'nunique')
).reset_index()

# 3. Tambahkan konversi Ton untuk mempermudah grafik dan pembacaan
df_tren_harian['Total_Tonase_Ton'] = (df_tren_harian['Total_Tonase_Kg'] / 1000).round(2)
df_tren_harian['Rata_Rata_Muatan_Per_Trip_Ton'] = (df_tren_harian['Rata_Rata_Muatan_Per_Trip'] / 1000).round(2)

# Urutkan berdasarkan urutan tanggal kronologis
df_tren_harian = df_tren_harian.sort_values(by='TANGGAL').reset_index(drop=True)

print("✔ Berhasil menghitung statistik tren harian periode Juni 2026.")
print(">>> Ringkasan Data Tren Harian (Poin 4.0):")
display(df_tren_harian)

# =====================================================================
# INTERPRETASI & IDENTIFIKASI TANGGAL KRITIKAL
# =====================================================================
print("\n=== ANALISIS PRODUKTIVITAS HARIAN ===")
hari_puncak_tonase = df_tren_harian.loc[df_tren_harian['Total_Tonase_Kg'].idxmax()]
hari_puncak_trip = df_tren_harian.loc[df_tren_harian['Total_Ritase_Trip'].idxmax()]

print(f"📌 Volume Sampah Tertinggi : Tanggal {hari_puncak_tonase['TANGGAL']} dengan total {hari_puncak_tonase['Total_Tonase_Ton']} Ton.")
print(f"📌 Ritase Perjalanan Terbanyak: Tanggal {hari_puncak_trip['TANGGAL']} dengan total {hari_puncak_trip['Total_Ritase_Trip']} Trip.")

# 4. EKSPOR JAWABAN NOMOR 4 KE EXCEL
nama_file_4 = 'Laporan_Tren_Harian_Poin4.xlsx'
df_tren_harian.to_excel(nama_file_4, index=False, sheet_name='Tren Harian Juni 2026')
print(f"\n🎉 Berkas Jawaban Poin 4.0 Berhasil Diekspor: '{nama_file_4}'")
```

5. Analisis Waktu Pelayanan Jembatan Timbang

```python
# =====================================================================
# JAWABAN POIN 5.0: ANALISIS EFISIENSI WAKTU OPERASIONAL (DURATION TIME)
# =====================================================================

print("=== PROSES PENYUSUNAN JAWABAN POIN 5.0 ===\n")

# 1. Identifikasi nama kolom Waktu/Jam secara fleksibel
col_masuk = [c for c in df_master_total.columns if 'MASUK' in c or 'JAM_1' in c or 'TIMBANG1' in c]
col_keluar = [c for c in df_master_total.columns if 'KELUAR' in c or 'JAM_2' in c or 'TIMBANG2' in c]

g_masuk = col_masuk[0] if col_masuk else 'JAM_MASUK'
g_keluar = col_keluar[0] if col_keluar else 'JAM_KELUAR'

if g_masuk in df_master_total.columns and g_keluar in df_master_total.columns:
    # 2. Konversi kolom jam menjadi format datetime agar bisa dikurangkan
    # (errors='coerce' akan mengubah data tidak valid menjadi NaT/Kosong)
    df_master_total['WAKTU_MASUK_DT'] = pd.to_datetime(df_master_total[g_masuk], format='%H:%M:%S', errors='coerce')
    df_master_total['WAKTU_KELUAR_DT'] = pd.to_datetime(df_master_total[g_keluar], format='%H:%M:%S', errors='coerce')

    # 3. Hitung selisih durasi dalam satuan Menit
    df_master_total['DURASI_MENIT'] = (df_master_total['WAKTU_KELUAR_DT'] - df_master_total['WAKTU_MASUK_DT']).dt.total_seconds() / 60

    # Filter durasi yang tidak masuk akal (misal: durasi negatif atau lebih dari 5 jam akibat salah input)
    df_waktu_valid = df_master_total[(df_master_total['DURASI_MENIT'] >= 0) & (df_master_total['DURASI_MENIT'] <= 300)]

    # 4. AGREGASI 1: Rata-rata Durasi Pelayanan per Kecamatan
    df_durasi_kec = df_waktu_valid.groupby('Kecamatan').agg(
        Total_Aktivitas_Trip=('NOPIN', 'count'),
        Rata_Durasi_Pelayanan_Menit=('DURASI_MENIT', 'mean')
    ).reset_index().round(2)

    df_durasi_kec = df_durasi_kec.sort_values(by='Rata_Durasi_Pelayanan_Menit', ascending=False).reset_index(drop=True)

    print(">>> 1. Rata-rata Waktu Pelayanan di Jembatan Timbang Berdasarkan Wilayah Kecamatan:")
    display(df_durasi_kec)

    # 5. AGREGASI 2: Identifikasi Jam Puncak Kepadatan (Berdasarkan Jam Masuk)
    df_waktu_valid['JAM_INPUT'] = pd.to_datetime(df_waktu_valid[g_masuk], format='%H:%M:%S', errors='coerce').dt.hour
    df_jam_puncak = df_waktu_valid.groupby('JAM_INPUT').size().reset_index(name='Jumlah_Truk_Masuk')
    df_jam_puncak = df_jam_puncak.sort_values(by='Jumlah_Truk_Masuk', ascending=False).reset_index(drop=True)

    print("\n>>> 2. Jam-Jam Puncak Kepadatan Operasional Jembatan Timbang:")
    display(df_jam_puncak.head(5))

    # 6. EKSPOR JAWABAN NOMOR 5 KE EXCEL
    nama_file_5 = 'Laporan_Efisiensi_Waktu_Poin5.xlsx'
    with pd.ExcelWriter(nama_file_5) as writer:
        df_durasi_kec.to_excel(writer, sheet_name='Durasi per Kecamatan', index=False)
        df_jam_puncak.to_excel(writer, sheet_name='Jam Puncak Kepadatan', index=False)
    print(f"\n🎉 Berkas Jawaban Poin 5.0 Berhasil Diekspor: '{nama_file_5}'")

else:
    print("⚠ Kolom penanda Jam Masuk / Jam Keluar tidak ditemukan pada dokumen.")
    print("Harap pastikan nama kolom jam timbang Anda di Excel (Contoh: JAM MASUK, JAM KELUAR).")
```
